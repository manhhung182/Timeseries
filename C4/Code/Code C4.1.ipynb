#### Dữ liệu GDP và Unemployment Rate (độ tuôi 14-64) của Mỹ

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import boxcox
```

```python
# Tải dữ liệu (US) GDP và Unemployment rate (14-64 tuổi)
df = pd.read_csv('../Data/GDP-UNEMP.csv',
                 parse_dates = ['DATE']).set_index('DATE')
df['GDP'] = boxcox(df['GDP'], lmbda = 0)
df.head()
```

```python
# Vẽ biểu đồ
fig, ax = plt.subplots(2,1,figsize=(12,5))
ax[0].plot(df['GDP'], color="navy", label = "GDP")
ax[1].plot(df['UNEMP'], color="navy", label = "Tỷ lệ thất nghiệp")
ax[0].legend()
ax[1].legend()
plt.show()
```

```python
# Tăng trưởng GDP và sự thay đổi của Unemployment rate
data = df.copy()
data = data.diff().dropna()
data.head()
```

```python
# Vẽ biểu đồ
fig, ax = plt.subplots(2,1,figsize=(12,6))
ax[0].plot(data['GDP'], color="navy", label = "Tăng trưởng GDP")
ax[1].plot(data['UNEMP'], color="navy", label = "Thay đổi tỉ lệ thất nghiệp")
ax[0].legend()
ax[1].legend()
plt.show()
```

```python
# Đồ thị điểm dữ liệu
plt.scatter(data["GDP"], data["UNEMP"], marker = "o", color="navy", alpha=0.5)
plt.xlabel("Tăng trưởng GDP")
plt.ylabel("Thay đổi tỉ lệ thất nghiệp")
```
