#### Đường trung bình trượt - Dữ liệu độ nhớt hoá chất

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Tải dữ liệu độ nhớt hoá chất
viscos = pd.read_csv('../Data/viscosity.csv').set_index('Time Period')
viscos.tail()
```

```python
# Biến đổi trung bình trượt
k=20    # Chu kỳ đường MA
rolling = viscos['Reading'].rolling(window=k) 
name = f'MA{k}'
viscos[name] = rolling.mean()

# Biểu đồ dữ liệu độ nhớt và đường MA
fig, ax = plt.subplots(figsize=(12,5))
ax.plot(viscos.Reading, 'o-', color='navy', linewidth=1.5, label="Độ nhớt")
ax.plot(viscos[name], 'd-', color='orangered', label=name)
plt.legend()
plt.show()
```
