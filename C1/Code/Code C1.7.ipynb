#### Hiệu chỉnh mùa - Dữ liệu doanh số đồ uống

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Dữ liệu doanh số bán lẻ đồ uống
retail = pd.read_csv('../Data/beveragesales.csv', 
                    parse_dates=['DATE']).set_index('DATE')[:96]

# Sai phân mùa và sai phân thường
fig, ax = plt.subplots(1,2,figsize=(12,4))
ax[0].plot(retail.diff(periods=12).dropna().values, 
           'o-', color='navy', linewidth=1.5, label="khử mùa")
ax[1].plot(retail.diff(periods=12).diff().dropna().values, 
           'o-', color='navy', linewidth=1.5, label='khử mùa và xu thế')
ax[0].legend(), ax[1].legend()
```
