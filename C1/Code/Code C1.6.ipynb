#### Hiệu chỉnh xu thế bằng sai phân

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Tải dữ liệu tỉ giá
exrate = pd.read_csv('../Data/eurusd.csv', parse_dates=['Date'],
                 infer_datetime_format=True)
exrate = exrate[['Date','Close']].set_index('Date')
```

```python
# Sai phân dữ liệu tỉ giá Eur/Usd
data = exrate.loc['2020':'2022',['Close']]
dif = data.Close.diff().dropna()

# Vẽ đồ thị
fig, ax = plt.subplots(figsize=(12,5))
ax.plot(dif, 'o-', color='navy', linewidth=1.5, label="sai phân dữ liệu tỉ giá")
ax.legend()
```
