#### Mô phỏng nhiễu trắng

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
```

```python
# Sinh nhiễu trắng Gauss
n = 100
sigma = 3
t = np.arange(n)
at = np.random.normal(0,sigma,n)

# Vẽ đồ thị đường
fig, ax = plt.subplots(figsize=(12,4))
ax.plot(t, at, 'o-', color='navy', linewidth=1.5, 
        label='Nhiễu trắng a(t) ~ WN(0,9)')
ax.legend()

# Vẽ lược đồ tự tương quan
fig = plot_acf(at, lags=30, zero = False)
plt.show()
```
