#### Phân tích các thành phần chuỗi thời gian

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
retail = pd.read_csv('../Data/beveragesales.csv', 
                    parse_dates=['DATE']).set_index('DATE')[:96]
retail
```

```python
# Phân tích thành phần của dữ liệu về doanh số bán đồ uống
from statsmodels.tsa.seasonal import seasonal_decompose

result = seasonal_decompose(retail, model='additive')
fig = result.plot()
fig.set_size_inches((10, 8))
fig.tight_layout()
plt.show()
```
