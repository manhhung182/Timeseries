#### Biến đổi boxcox - Dữ liệu số vết đen mặt trời

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from scipy.stats import boxcox

# Tải dữ liệu số điểm đen mặt trời
sunspot = pd.read_excel('../Data/sunspot.xls').set_index('Year')

# Biến đổi logarit và tối ưu tham số biến đổi
sunspot['log'] = boxcox(sunspot['Total sunspot number'], lmbda=0)
sunspot['boxcox'],lam = boxcox(sunspot['Total sunspot number'], lmbda=None)
print('(optimized) lambda = ', lam)

# Vẽ biểu đồ
fig, ax = plt.subplots(1,2,figsize=(14,5))
ax[0].plot(sunspot.log, 'o-', color='navy', linewidth=1.5,label="log transform")
ax[1].plot(sunspot.boxcox, 'o-', color='navy', linewidth=1.5, 
           label="optimized boxcox transform")
ax[0].legend() , ax[1].legend()
```
