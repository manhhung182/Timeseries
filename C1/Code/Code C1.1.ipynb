#### Biểu đồ đường - Dữ liệu tỉ giá Eur/usd

```python
# Khai báo thư viện
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Tải dữ liệu tỉ giá
exrate = pd.read_csv('../Data/eurusd.csv', parse_dates=['Date'],
                 infer_datetime_format=True)
exrate = exrate[['Date','Close']].set_index('Date')
exrate.head()
```

```python
# Biểu đồ dữ liệu tỉ giá Eur/usd
fig, ax = plt.subplots(figsize=(12,5))
ax.plot(exrate.Close.loc['2020':'2022'], 
        '-', 
        color='navy',  
        linewidth = 2.0,
        label = "Tỉ giá Eur/Usd")
plt.legend()
plt.show()
```
