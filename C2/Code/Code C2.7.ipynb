#### Mô phỏng hàm ACF, PACF của quá trình ARMA(2,1)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
# Sinh dữ liệu mô phỏng quá trình ARMA
n =100
a = np.random.normal(0,5,n) 
y = np.zeros(n)
for t in range(2,n):
    y[t] = 0.4*y[t-1] + 0.5*y[t-2] + a[t] - 0.8*a[t-1]  

# Vẽ đồ thị ACFs, PACFs
ACF, ax = plt.subplots(1,2, figsize = (12,4))
fig = plot_acf(y, lags = 25, zero = False, 
               ax = ax[0], title = 'Biểu đồ ACF của y(t)')
fig = plot_pacf(y, lags = 25, zero = False, method = 'ywm',
                ax = ax[1], title = 'Biểu đồ PACF của y(t)')
```
