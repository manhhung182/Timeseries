#### Mô phỏng quá trình MA(2)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
# Sinh dữ liệu mô phỏng quá trình MA(2)
n =200
a = np.random.normal(0,5,n) # nhiễu trắng Gauss 
y = np.zeros(n)
for t in range(2,n):
    y[t] = a[t]  + 0.4*a[t-1] + 0.5*a[t-2]  

# Vẽ đồ thị đường
MA2, ax = plt.subplots(figsize=(12,4))
plt.plot(y, '-o', linewidth=1.5)

# Vẽ biểu đồ ACFs
ACF, ax = plt.subplots()
fig = plot_acf(y, lags = 25, zero = False, ax = ax)
ax.set(title = 'Biểu đồ ACF')
```
