#### Mô phỏng quá trình AR(1) dừng

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
# Sinh ngẫu nhiên quá trình AR(1) dừng
n = 100
a = np.random.normal(0,5,n) 
y = np.zeros(n)
z = np.zeros(n)
for  t in range(1,n):
    y[t] = 0.7 * y[t-1] + a[t]
    z[t] = -0.7 * z[t-1] + a[t]
    
# Vẽ đồ thị đường
AR1, ax1 = plt.subplots(1,2, figsize = (12,4))
ax1[0].plot(y, '-o', linewidth=1.5)
ax1[1].plot(z, '-o', linewidth=1.5)
ax1[0].set(title = 'y(t)')
ax1[1].set(title = 'z(t)' )

# Vẽ lược đồ tương quan mẫu
ACF, ax2 = plt.subplots(1,2, figsize = (12,4))
ACF = plot_acf(y, lags = 25, zero = False, ax = ax2[0])
ACF = plot_acf(z, lags = 25, zero = False, ax = ax2[1])
ax2[0].set(title = 'Biểu đồ ACF của y(t)')
ax2[1].set(title = 'Biểu đồ ACF của z(t)')
plt.show()
```
