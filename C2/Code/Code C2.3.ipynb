#### Mô phỏng hàm ACF của quá trình AR(2)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
# Sinh dữ liệu quá trình AR(2)
n =100
a = np.random.normal(0,5,n)
y = np.zeros(n)
z = np.zeros(n)
for t in range(2,n):
    y[t] = 0.5*y[t-1] + 0.1875*y[t-2] + a[t] 
    z[t] = z[t-1] - 0.5*z[t-2] + a[t]

# Vẽ đồ thị đường
AR2, ax1 = plt.subplots(1,2, figsize = (12,4))
ax1[0].plot(y, '-o', linewidth=1.5)
ax1[1].plot(z, '-o', linewidth=1.5)
ax1[0].set(title = 'y(t)')
ax1[1].set(title = 'z(t)')

# Vẽ lược đồ tự tương quan
ACF, ax2 = plt.subplots(1,2, figsize = (12,4))
fig = plot_acf(y, lags = 25, zero = False, ax = ax2[0])
fig = plot_acf(z, lags = 25, zero = False, ax = ax2[1])
ax2[0].set(title = "ACF của quá trình y(t)")
ax2[1].set(title = "ACF của quá trình z(t)")
```
