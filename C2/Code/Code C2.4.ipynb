#### Tính các hệ số PACFs của quá trình AR(2)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
# Sinh dữ liệu của quá trình AR
n = 100
a = np.random.normal(0,5,n) 
y = np.zeros(n)
z = np.zeros(n)
for t in range(2,n):
    y[t] = 0.6*y[t-1] + 0.25*y[t-2] + a[t] 
    z[t] = (5/4)*z[t-1] - (1/2)*z[t-2] + (1/16)*z[t-3] + a[t] 
    
# Vẽ biểu đồ PACFs
PACF, ax = plt.subplots(1,2, figsize = (12,4))
fig = plot_pacf(y, lags = 25, zero = False, ax = ax[0],
                method= 'ywm', title = "PACF của quá trình y(t)")
fig = plot_pacf(z, lags = 25, zero = False, ax = ax[1],
                method= 'ywm', title = "PACF của quá trình z(t)")
```
