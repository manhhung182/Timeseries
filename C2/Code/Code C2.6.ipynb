#### Mô phỏng quá trình ARMA(1,1)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
# Sinh dữ liệu mô phỏng cho quá trình ARMA
n =100
sigma = 5 # std = 5, var = 25
a = sigma*np.random.randn(n) # Sinh ra nhiễu trắng WN(0, 25) 
y = np.zeros(n)
for t in range(1,n):
    y[t] = 0.7*y[t-1] + a[t] + 0.6*a[t-1]  

# Vẽ đồ thị đường
ARMA11, ax1 = plt.subplots(figsize=(12,4))
ax1.plot(y, '-o', linewidth=1.5)

# Vẽ hàm ACFs, PACFs của y(t)
ACFy, ax2 = plt.subplots(1,2, figsize = (12,4))
fig = plot_acf(y, lags = 25, zero = False, 
               ax = ax2[0], title = 'Biểu đồ ACF của y(t)')
fig = plot_pacf(y, lags = 25, zero = False, method = 'ywm',
                ax = ax2[1], title = 'Biểu đồ PACF của y(t)')
```
