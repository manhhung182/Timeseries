#### Phân tích dữ liệu tỉ giá Euro-Usd

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
# Tải dữ liệu từ file
exrate = pd.read_csv('../Data/eurusd.csv', parse_dates = ['Date'],
                     infer_datetime_format=True)
exrate = exrate[['Date','Close']].set_index('Date')
exrate.head()
```

```python
# Biểu đồ dữ liệu tỉ giá
plt.subplots(figsize=(12,5))
plt.plot(exrate, color="navy", label = "Tỉ giá Eur/Usd")
plt.legend()
```

```python
# Vẽ biểu đồ ACF
fig = plot_acf(exrate, lags = 30, zero = False)
```

```python
# Lấy sai phân chuỗi
diff = exrate.diff().dropna()

plt.subplots(figsize=(12,5))
plt.plot(diff, color="navy", label = "Chuỗi sai phân")
plt.legend()

fig = plot_acf(diff, lags = 50, zero = False)
```
